```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')

test_data.head()
```

```python
con1 = (train_data.Y_01 >= 0.2) & (train_data.Y_01 <= 2)
con2 = (train_data.Y_02 >= 0.2) & (train_data.Y_02 <= 2.1)
con3 = (train_data.Y_03 >= 0.2) & (train_data.Y_03 <= 2.1)
con4 = (train_data.Y_04 >= 7) & (train_data.Y_04 <= 19)
con5 = (train_data.Y_05 >= 22) & (train_data.Y_05 <=36.5)
con6 = (train_data.Y_06 >= -19.2) & (train_data.Y_06 <= 19)
con7 = (train_data.Y_07 >= 2.4) & (train_data.Y_07 <= 4)
con8 = (train_data.Y_08 >= -29.2) & (train_data.Y_08 <= -24)
con9 = (train_data.Y_09 >= -29.2) & (train_data.Y_09 <= -24)
con10 = (train_data.Y_10 >= -30.6) & (train_data.Y_10 <= -20)
con11 = (train_data.Y_11 >= 19.6) & (train_data.Y_11 <= 26.6)
con12 = (train_data.Y_12 >= -29.2) & (train_data.Y_12 <= -24)
con13 = (train_data.Y_13 >= -29.2) & (train_data.Y_13 <= -24)
con14 = (train_data.Y_14 >= -29.2) & (train_data.Y_14 <= -24)

train_data = train_data.loc[con1&con2&con3&con4&con5&con6&con7&con8&con9&con10&con11&con12&con13&con14,:]
train_data
```

```python
train_data_X = train_data.filter(regex='X_')
train_data_y = train_data.filter(regex='Y_')
train_data_y
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
train_data_X = scaler.fit_transform(train_data_X)
test_data = scaler.transform(test_data)
```

```python
from sklearn.model_selection import KFold, train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.multioutput import MultiOutputRegressor
from sklearn.metrics import mean_squared_error

kfold = KFold(n_splits=5, shuffle=True, random_state=2022)
lr = LinearRegression()

train_data_X = np.array(train_data_X)
train_data_y = np.array(train_data_y)

RMSE_history = []

for train_index, test_index in kfold.split(train_data_X):
    X_train, X_test = train_data_X[train_index], train_data_X[test_index]
    y_train, y_test = train_data_y[train_index], train_data_y[test_index]
    
    MultiOutputRegressor(lr.fit(X_train,y_train))
    y_pred = lr.predict(X_train)
    mse = mean_squared_error(y_train, y_pred)
    rmse = np.sqrt(mse)
    RMSE_history.append(rmse)
    
print(RMSE_history)
print(np.mean(RMSE_history))
    
```

```python
test_data_y = lr.predict(test_data)
test_data_y
```

```python
test_data_y = pd.DataFrame(test_data_y,columns = train_data.filter(regex='Y_').columns)
submission = pd.concat([test_ID, test_data_y],axis = 1)
submission
```
